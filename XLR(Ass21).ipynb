```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer

df = pd.read_csv('Civil_Engineering_Regression_Dataset.csv')

print(df.head())
independent_vars = df.drop(columns=['Construction_Cost'])
dependent_var = df['Construction_Cost']
missing_values = df.isna().sum()
print("Missing Values Count:")
print(missing_values)
df.fillna(df.median(), inplace=True)
df.drop_duplicates(subset=['Project_ID'], keep='first', inplace=True)
print(df.describe())
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
df.to_csv('Day_21_Civil_Engineering_Regression_Dataset.csv', index=False)
```
